```python
import pandas as pd

# Load your CSV file into a DataFrame
df = pd.read_csv('/content/drive/MyDrive/CSC 302 Final Project/Spotify Top 100 USA July 4.csv')

# Display the first few rows to understand its structure
print(df.head())
```

```python
# Count occurrences of each artist
artist_counts = df['Artist'].value_counts().reset_index()
artist_counts.columns = ['Artist', 'count']

# Display the counts
print(artist_counts)
```

```python
import plotly.express as px

# Create a tree map using Plotly with custom colors
fig = px.treemap(artist_counts,
                 path=['Artist'],
                 values='count',
                 title='Artist Occurrences In Top 100(Week of July 4, 2024 - USA)',
                 color='count',  # Use 'count' for color coding
                 color_continuous_scale='Cividis')
fig.update_layout(title_x=0.53)
# Show the plot
fig.show()
```
